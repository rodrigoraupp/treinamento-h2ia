Mas vamos responder uma pergunta espec√≠fica: Qual foi a taxa geral de sobreviv√™ncia?

## Investiga√ß√£o 1: Mulheres e Crian√ßas Primeiro?

Existe uma regra antiga do mar que diz que mulheres e crian√ßas devem ser salvas primeiro. Ser√° que isso foi respeitado no caos do Titanic?

Para descobrir, precisamos filtrar os dados. No Pandas, fazemos perguntas l√≥gicas aos dados.

Vamos criar dois novos DataFrames: um s√≥ com homens e outro s√≥ com mulheres.

```python
# verificando dados √∫nicos na coluna Sex
df['Sex'].value_counts()
```

```python
# Filtrando Mulheres
women = df[df['Sex'] == 'female']

# Fa√ßa a filtragem de homens
men = df.query('Sex == "male"')

print(f"Total de Mulheres a bordo: {len(women)}")
print(f"Total de Homens a bordo: {len(men)}")
```

<details>
<summary><strong>üïµÔ∏è‚Äç‚ôÄÔ∏è Clique aqui para entender a m√°gica dos colchetes []</strong></summary>
<br>
Voc√™ deve estar se perguntando: <em>"Por que repetimos o <code>df</code> dentro do <code>df</code>?"</em>
<br><br>
A linha <code>df[df['Sex'] == 'female']</code> acontece em dois passos:
<ol>
  <li><strong>A Pergunta (O Miolo):</strong> O Python executa primeiro o que est√° dentro: <code>df['Sex'] == 'female'</code>. Isso n√£o retorna dados, mas sim uma lista de <code>True</code> (Verdadeiro) ou <code>False</code> (Falso) para cada linha. √â como se ele passasse em cada passageiro perguntando "Voc√™ √© mulher?".</li>
  <li><strong>O Filtro (Os Colchetes de fora):</strong> Quando passamos essa lista de Verdadeiro/Falso para o <code>df[...]</code>, o Pandas entende o seguinte: "Onde for <strong>True</strong>, eu mantenho a linha. Onde for <strong>False</strong>, eu jogo fora".</li>
</ol>
Isso √© chamado de <strong>M√°scara Booleana</strong> (Boolean Mask). Se voc√™ quiser ver o passo a passo, deixei um c√≥digo escondido que mostra o funcionamento desse processo.
</details>

```python
# @title
# --- RAIO-X DO C√ìDIGO ---
# Vamos ver o que acontece por baixo do cap√¥

mask = df['Sex'] == 'female'

print("O que a 'm√°scara' cont√©m:")
print(mask.head())
print("\n----------------\n")

# Passo 2: Aplicando a m√°scara
print("Usando a m√°scara para filtrar:")
resultado = df[mask]
resultado.head()
```

<details>
<summary><strong>‚ö†Ô∏è Dica: O erro mais comum nesse passo</strong></summary>
<br>
Se voc√™ receber um erro de sintaxe, verifique se usou <code>==</code> (dois iguais).
<br><br>
<ul>
    <li><code>=</code> (um igual) serve para <strong>definir</strong> valor (ex: <code>x = 10</code>).</li>
    <li><code>==</code> (dois iguais) serve para <strong>comparar</strong> valor (ex: <code>x √© igual a 10?</code>).</li>
</ul>
Dentro dos filtros, estamos sempre <strong>comparando</strong>, ent√£o use sempre <code>==</code>.
</details>

Agora que separamos os grupos, vamos calcular a m√©dia de sobreviv√™ncia de cada um.

```python
# Calcule a m√©dia da coluna 'Survived' para mulheres e homens
mean_women = women['Survived'].mean() # IMPLEMENTE usando o dataframe women, selecionando a coluna survived e tirando a m√©dia
mean_men = men['Survived'].mean() # IMPLEMENTE usando o dataframe men, selecionando a coluna survived e tirando a m√©dia

print(f"Chance de sobreviv√™ncia (Mulheres): {mean_women*100:.1f}%")
print(f"Chance de sobreviv√™ncia (Homens): {mean_men*100:.1f}%")
```

O que podemos tirar de conclus√£o?

A m√©dia de sobreviv√™ncia das mulheres que estavam √† bordo do Titanic √© substancialmente maior que a m√©dia masculina. A probabilidade de uma mulher ser sobrevivente √© maior que a de um homem.

## Investiga√ß√£o 2: O Dinheiro Compra a Vida?

E quanto √† classe social? Ser√° que os ricos da 1¬™ classe tiveram prioridade nos botes em rela√ß√£o aos imigrantes da 3¬™ classe?

* Para isso, n√£o precisamos criar v√°rios filtros manuais. Podemos usar o poderoso `.groupby()`.
* Ele pega os dados, separa em grupos baseados em uma coluna (neste caso `Pclass`) e aplica uma conta (m√©dia).

```python
# verificando valores √∫nicos da coluna Pclass
df['Pclass'].value_counts()
```

```python
# Agrupando por Classe e calculando a m√©dia de sobreviv√™ncia
survival_by_class = df.groupby('Pclass')['Survived'].mean()

print(survival_by_class)
```

Classe social foi importante aqui? Ou tanto fez?

Quem pagou mais pelo bilhete de viagem teve maior probabilidade de sobreviver ao naufr√°gio. A terceira classe foi a que mais teve √≥bitos.

## Visualizando os Dados

N√∫meros s√£o √≥timos, mas gr√°ficos contam hist√≥rias. Vamos usar o matplotlib para visualizar essa desigualdade entre as classes.

Vamos criar um gr√°fico de barras simples.

```python
# Plotando o gr√°fico de barras
survival_by_class.plot(kind='bar', color=['gold', 'gray', 'brown'])

plt.title('Sobreviv√™ncia por Classe Social')
plt.ylabel('Chance de Sobreviver')
plt.xlabel('Classe (1=Rico, 3=Pobre)')
plt.xticks(rotation=0) # Deixa os n√∫meros do eixo X retos
plt.show()
```
